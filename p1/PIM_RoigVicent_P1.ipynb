```python
import time
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from skimage import io
from scipy import misc, ndimage
```

## Semana #1. Imagen hibrida en el dominio del espacio

```python
#Load images
img1 = Image.open("./img/human.png")
img2 = Image.open("./img/cat.png")

#Resize
size1=int(round(img1.size[0]*(4/3.)));
size2=int(round(img1.size[1]*(4/3.)));
img1 = img1.resize((size1,size2), Image.ANTIALIAS)

#Crop
left = 50
top = 50
right = 51+img2.size[0]-1
bottom = 51+img2.size[1]-1
img1=img1.crop((left, top, right, bottom))

#Visualization
plt.figure(1)
plt.subplot(1,2,1) 
imgplot1=plt.imshow(img1)
plt.subplot(1,2,2)
imgplot1=plt.imshow(img2)
```

```python
# * * * * * * * * * *
# * Gaussian Filter *
# * * * * * * * * * *

def gaussianFilter(sigma):
    SZ = 6*sigma + 1
    radius = (SZ-1)/2
    [xx,yy]=np.meshgrid(np.linspace(-radius,radius),np.linspace(-radius,radius))
    gaussian = np.exp(-0.5*(xx*xx+yy*yy)/(sigma*sigma))
    
    return gaussian/np.sum(gaussian)
```

```python
g = gaussianFilter(9)
print '> Filter size:', g.shape
print '> Gaussian max:', g.max()
print '> Gaussian sum:', np.sum(g)

#Visualization
plt.imshow(g)
```

```python
# * * * * * * * * * * * * * * * * * * *
# * custom Spatial-Domain Convolution *
# * * * * * * * * * * * * * * * * * * *

def rgb_convolve(I, F):
    res = np.zeros(I.shape)
    for i in range(3):
        res[:,:,i] = ndimage.convolve(I[:,:,i], F)
    return res

'''image uint8 normalize function'''
def norm(img):
    return np.real((img - np.min(img)) * 256 / (np.max(img) - np.min(img)))
```

```python
img1 = np.array(img1).astype('double')/255.
img2 = np.array(img2).astype('double')/255.

# LowPass Filtering:
gaussian = gaussianFilter(9)
t0 = time.time()
img1_lp = rgb_convolve(img1, gaussian)
img2_lp =  rgb_convolve(img2, gaussian)
elapsed = time.time() - t0

# HighPass Filtering:
'''An easy way of making a highpass filter is to simply subtract a lowpass filtered image from the original.'''
img2_hp = img2 - img2_lp

# Final Hybrid Image as a comb. of both LP & HP computed 
mix = (img2_hp + img1_lp)/2.
```

```python
print "Time taken: ", elapsed, "seconds."

#Visualization
plt.figure(1)
plt.subplot(1,3,1)
plt.imshow(norm(img1_lp).astype('uint8'))
plt.title('Human Lowpass Filt.', size=10)

plt.subplot(1,3,2)
plt.imshow(norm(img2_hp).astype('uint8'))
plt.title('Cat Highpass Filt.', size=10)

plt.subplot(1,3,3)
plt.imshow(norm(mix).astype('uint8'))
plt.title('Hybrid Image', size=10)

plt.gcf().set_size_inches((14,14))
```

## Semana #2. Imagen hibrida en el dominio de la frecuencia

```python
# * * * * * * * * * * * * * * * * * * * *
# * custom Frequency Domain Convolution *
# * * * * * * * * * * * * * * * * * * * * 
from scipy import fftpack

def rgb_convolveFFT(I, F, SZ=1024):
    fil_fft = fftpack.fft2(F, (SZ, SZ)) #fft filter and pad to the same size as the image
    hs = np.floor(F.shape[0]/2.)
    
    res = np.zeros(I.shape)
    for i in range(3):
        im_fft = fftpack.fft2(I[:,:,i], (SZ, SZ))
        im_fil_fft = im_fft * fil_fft  # Multiply both fft images element-wise
        im_fil = np.real(fftpack.ifft2(im_fil_fft)) # Inverse fft
        res[:,:,i] = im_fil[hs:I.shape[0] + hs, hs:I.shape[1] + hs] # Remove padding

    return res


def img_comparison(img1, img2):
    dif = np.sum(abs(img2 - img1)) / float(img1.shape[0] * img1.shape[1] * img1.shape[2])
    print '> % diff:\t' + str(dif/256.)
```

```python
# * * * * * * * * * * * * * * * * * * *
# * Filtering in the Frecuency Domain *
# * * * * * * * * * * * * * * * * * * *

# Transform the original image. Fftsize should be order of 2 and include padding.
fftsize = 1024

# Create the filter mask in the spatial domain
g = gaussianFilter(9)

# lowpass filtering
t0 = time.time()
img1_lp_fft = rgb_convolveFFT(img1, g)
img2_lp_fft = rgb_convolveFFT(img2, g, fftsize)
elapsed_fft = time.time() - t0

# HighPass Filtering:
'''An easy way of making a highpass filter is to simply subtract a lowpass filtered image from the original.'''
img2_hp_fft = img2 - img2_lp_fft

# Final Hybrid Image as a comb. of both LP & HP computed 
mix_fft = (img2_hp_fft + img1_lp_fft)/2. 
```

```python
print "Time taken: ", elapsed_fft, "seconds."

#Visualization
plt.figure(1)
plt.subplot(1,3,1)
plt.imshow(norm(img1_lp_fft).astype('uint8'))
plt.title('Human Lowpass Filt.', size=10)

plt.subplot(1,3,2)
plt.imshow(norm(img2_hp_fft).astype('uint8'))
plt.title('Cat Highpass Filt.', size=10)

plt.subplot(1,3,3)
plt.imshow(norm(mix_fft).astype('uint8'))
plt.title('Hybrid Image', size=10)

plt.gcf().set_size_inches((14,14))
```

## Image Result Comparison:

```python
#Verify Image Properties
print "[Spatial D. Highpass Fil. / Frequency D. Highpass Fil.]"
img_comparison(img2_hp, img2_hp_fft)
print "\n[Spatial D. Lowpass Fil. / Frequency D. Lowpass Fil.]"
img_comparison(img1_lp, img1_lp_fft)
print "\n[Final Hybrid Difference]"
img_comparison(mix, mix_fft)

print "\n[Spatial D. Fil. VS Frequency D. Fil. ELAPSED TIME ]"
print "-> Spatial Dom.  \t", elapsed, "sec." 
print "-> Frequency Dom.\t", elapsed_fft, "sec."

# Face to Face Visualization
fig = plt.figure(1)

plt.subplot(3,2,1)
plt.imshow(norm(img2_hp).astype('uint8'))
plt.title('Highpass', size=12)
plt.subplot(3,2,2)
plt.imshow(norm(img2_hp_fft).astype('uint8'))
plt.title('iFFT Highpass', size=12)

plt.subplot(3,2,3)
plt.imshow(norm(img1_lp).astype('uint8'))
plt.title('Lowpass', size=12)
plt.subplot(3,2,4)
plt.imshow(norm(img1_lp_fft).astype('uint8'))
plt.title('iFFT Lowpass', size=12)

plt.subplot(3,2,5)
plt.imshow(norm(mix).astype('uint8'))
plt.title('Hybrid', size=12)
plt.subplot(3,2,6)
plt.imshow(norm(mix_fft).astype('uint8'))
plt.title('iFFT Hybrid', size=12)

plt.gcf().set_size_inches((10,10))
fig.tight_layout()
```
