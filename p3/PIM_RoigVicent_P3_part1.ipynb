### Gradient Magnitude and Energy Matrix

Dynamic programming is used in seam carving for computing seams. We will store the results of sub-calculations in order to simplify calculating a more complex alg.

If attempting to compute a vertical seam (path) of lowest energy, for each pixel in a row we compute the energy of the current pixel plus the energy of one of the three possible pixels above it.

```python
# * * * * * * * * * * * * * * *
# * Gradient Magnitude Matrix *
# * * * * * * * * * * * * * * *
def gradient(img):
    '''
    compute image gradient magnitude matrix.
    '''
    gx, gy = np.gradient(img)
    grad = np.sqrt(gx*gx + gy*gy)
    return grad

# * * * * * * * * *
# * Energy Matrix *
# * * * * * * * * *
def compute_energy(im):
    '''
    This implements the dynamic programming seam-find algorithm. For an m*n picture, this algorithm
     takes O(m*n) time 
    '''
    im_width, im_height = im.shape
    cost = np.zeros(im.shape)
    cost[0] = im[0] #first row for energy matrix as the original gradient magnitude
  
    for x in range(1, im_width):
        for y in range(im_height):
            if y == 0:
                min_val = min( cost[x-1, y], cost[x-1, y+1] )
            elif y < im_height - 2:
                min_val = min( cost[x-1, y], cost[x-1, y+1] )
                min_val = min( min_val, cost[x-1, y-1] )
            else:
                min_val = min( cost[x-1, y], cost[x-1, y-1] )
            cost[x,y] = im[x,y] + min_val
    return cost
```

```python
grad = gradient(img)
energy = compute_energy(grad)

#Visualization
fig = plt.figure(2) 
plt.subplot(1,2,1)
plt.imshow(norm(grad), cmap="gray")
plt.title('Gradient Magnitude')
plt.axis('off')
plt.subplot(1,2,2)
plt.imshow(norm(energy), cmap="gray" )
plt.title('Energy')
plt.axis('off')
plt.gcf().set_size_inches((20,20))
fig.tight_layout()
plt.show()
```

### Finding and Removing Vertical Seams

Computing the seam consists of finding the path of minimum energy cost from one end of the image to another.

Seams can be either vertical or horizontal. A vertical seam is a path of pixels connected from top to bottom in an image with one pixel in each row.[1] A horizontal seam is similar with the exception of the connection being from left to right. The importance/energy function values a pixel by measuring its contrast with its neighbor pixels.

```python
def find_vertical_seam(im):
    '''
    Takes a grayscale img and returns the lowest energy vertical seam as a list of pixels (2-tuples). 
    This implements the dynamic programming seam-find algorithm. For an m*n picture, this algorithm
    takes O(m*n) time 
    @im: a grayscale image
    '''
    assert len(im.shape) == 2
    
    im_width, im_height = im.shape
    grad = gradient(im)
    cost = compute_energy(grad) #recalculated energy function
    
    min_val = 1e1000
    path = []
    for y in range(im_height):
        if cost[im_width-1,y] < min_val:
            min_val = cost[im_width-1,y]
            min_ptr = y
            
    pos = (im_width-1, min_ptr)
    path.append(pos) 
    while pos[0] != 0:
        val = cost[pos] - grad[pos]
        x,y = pos
        if y == 0:
            if val == cost[x-1,y+1]:
                pos = (x-1,y+1) 
            else:
                pos = (x-1,y)
        elif y <= im_height - 2:
            if val == cost[x-1,y+1]:
                pos = (x-1,y+1) 
            elif val == cost[x-1,y]:
                pos = (x-1,y)
            else:
                pos = (x-1,y-1)
        else:
            if val == cost[x-1,y]:
                pos = (x-1,y)
            else:
                pos = (x-1,y-1) 
                
        path.append(pos)  
    #print "Reconstruction Complete."
    return path


def mark_seam (img, path, mark_as='red'):
    '''
    Marks a seam for easy visual checking
    @img: an input img
    @path: the seam
    '''
    assert mark_as in ['red','green','blue','black','white']
    assert len(img.shape) == 3
    
    #print "Marking seam..."
    for pixel in path:
        if mark_as == 'red':
           img[pixel] = (255,0,0)
        elif mark_as == 'green':
            img[pixel] = (0,255,0)
        elif mark_as == 'blue':
            img[pixel] = (0,0,255)
        elif mark_as == 'white':
            img[pixel] = (255,255,255)
        elif mark_as == 'black':
            img[pixel] = (0,0,0)
    #print "Marking Complete."
    
    return img;


def delete_vertical_seam (img, path):
    '''
    Deletes the pixels in a vertical path from img
    @img: an input img
    @path: pixels to delete in a vertical path
    '''
    #print "Deleting Vertical Seam..."
    img_height, img_width = img.shape[:2] 
    i = np.zeros((img_height, img_width-1, img.shape[2]))
                 
    path_set = set(path)
    seen_set = set()
    for x in range(img_height):
        for y in range(img_width):
            if (x,y) not in path_set and x not in seen_set:
                i[x,y] = img[x,y]
            elif (x,y) in path_set:
                seen_set.add(x)
            else:
                i[x,y-1] = img[x,y]
    #print "Deletion Complete."
    return i;
```

```python
# * * * * * * * * *
# * Seam Carving  *
# * * * * * * * * *
def seam_carving(it, RGB, verbose=False):
    '''
    After we have the energy of the image, we generate a list of seams. 
    We then remove the seams from the image, reducing the size of the image as a result
    '''
    seams = []
    rgb = RGB.copy();
    
    for i in range(it):
        gray = color.rgb2gray(rgb)
        path = find_vertical_seam(gray)
        
        rgb = mark_seam(rgb, path, 'blue')
        seams.append([plt.imshow(rgb.astype('uint8'))])
        rgb = delete_vertical_seam(rgb, path)
        
        if verbose:
            print "iteration "+str(i),"\tseam carved...", rgb.shape
        
    return rgb, seams;
```

### Execution Test and Results

```python
N = 100 #number of seams to be removed (iterations)
fig = plt.figure(3)
result, seams = seam_carving(N, RGB, True);
```
